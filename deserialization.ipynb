```python
from pydantic import BaseModel, ValidationError
```

```python
# A simple creation of Pydantic object

class Person(BaseModel):
    first_name: str
    last_name: str
    age: int

p = Person(first_name="sandeep", last_name="srivastava", age=30)
p
```

```python
# Deserialize (validate and deserialize) a Python Dictionary
data = {
    "first_name":"shlok",
    "last_name": "srivastava",
    "age": 14
}
p_from_dict = Person.model_validate(data)
p_from_dict
```

```python
# Deserialize (validate and deserialize) a Python Dictionary
# with a known validation issue

missing_data = {
    "first_name":"siona",
    "last_name": "srivastava"
}

try:
    p2_from_dict = Person.model_validate(missing_data)

except ValidationError as ex:
    #print(f"This program threw a validation error - {ex}") 
    print(f"This program quietly threw a validation error.") 
```

```python
# Deserialize (validate and deserialize) a JSON string
# with a known validation issue 

data_json = """
{
    "first_name":"siona",
    "last_name": "srivastava",
    "age" : "18"
}
"""
p_from_json = Person.model_validate(data_json)
p_from_json
```

### Type Coercion
Pydantic will type coerce when application.
A string of "50.00" will be coreced to int when assigned to named variable "age"

```python
pClass = Person(first_name="Shitiz", last_name="Sinha", age="50.00")
type(pClass.first_name), type(pClass.last_name), type(pClass.age)
```
